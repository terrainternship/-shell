# Разделение классов во всех файлах в папке

```python
source_folder = '/content/drive/MyDrive/GUNS/People with arms.v5i.yolov8/train/all_labels'
destination_folder = '/content/drive/MyDrive/GUNS/People with arms.v5i.yolov8/train/labels'
```

Изначальные классы:

0 class - human

1 class - human with gun

2 class - gun

Не берем класс 1 - human with gun

Необходимо сменить индекс класса 2 - gun, чтобы в итоге иметь следующие два класса:

0 class - human

1 class - gun

```python
# function to read a file from source and create another in the destination directory
def get_two_classes(filename, filepath, destination_folder):
    destination_filepath = os.path.join(destination_folder, filename)
    with open(filepath) as file_in:
        with open(destination_filepath, 'w') as file_out:       
            for line in file_in:
                if int(line[0]) == 0:
                    print(line.rstrip())
                    file_out.write(line) # строку с классом 0 записываем без изменений
                if int(line[0]) == 2:
                    new_line = '1' + line[1:] # заменяем индекс класса на 1
                    file_out.write(new_line) # записываем новую строку
                    print(new_line.rstrip())
```

```python
# iterate over files in the directory
for filename in os.listdir(source_folder):
    filepath = os.path.join(source_folder, filename)
    # checking if it is a file
    if os.path.isfile(filepath):
        get_two_classes(filename, filepath, destination_folder)
```
